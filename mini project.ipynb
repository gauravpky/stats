```python
#2.Analyze the data to provide the hypothesis testing conclusion. What is the p-value for your test? 
#What is your recommendation for Par Inc.?
#3(2)
stat,p=ttest_ind(golf['Current'],golf['New']) 
stat
p 
alpha=0.05
if p < alpha:
    print('reject')
else:
    print('accept') 
```

```python
#3. What is the 95% confidence interval for the population mean of each model, and what is the 95% 
#confidence interval for the difference between the means of the two population?
#3(3)
current=golf['Current']
cur_mean=current.mean()
print(cur_mean )   
cur_std=current.std()   
print(cur_std) 

from scipy.stats import t
t_bounds=t.interval(0.95,len(current)-1)   
print(t_bounds)

ci_curr=[(cur_mean)+ (i*(cur_std/len(current)**0.5)) for i in t_bounds]
print(cur_mean)
print(ci_curr)    

##########################################################################

new=golf['New']
new_mean=new.mean()
print(new_mean)   
new_std=new.std()   
print(new_std) 

from scipy.stats import t
t_bounds=t.interval(0.95,len(new)-1)
print(t_bounds)

ci_new=[(new_mean)+ (i*(new_std/len(new)**0.5)) for i in t_bounds]
print(new_mean)
print(ci_new)  
```
